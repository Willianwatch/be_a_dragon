```python
def create_data_set():
    group = np.array([[1.0, 1.1], [1.0, 1.0], [0, 0], [0, 0.1]])
    labels = np.array(['A', 'A', 'B', 'B']).T
    return group, labels
```

```python
from collections import Counter
```

```python
import heapq
```

```python
def classify(inX, data_set, labels, k):
    data_set_size = np.shape(data_set)[0]
    diff_mat = np.tile(inX, (data_set_size, 1)) - data_set
    sq_diff_mat = diff_mat**2
    sq_distances = np.sum(sq_diff_mat, axis=1)
    distances = sq_distances**0.5
    nearest_k = heapq.nsmallest(k, enumerate(sq_distances), key=lambda i: i[1])
    class_count = Counter(labels[i[0]] for i in nearest_k)
    return class_count.most_common(1)[0][0]
```

```python
group, labels = create_data_set()
```

```python
classify([0, 0], group, labels, 3)
```

```python
def file_to_matrix(file_name):
    data_set = []
    degrees = []
    with open(file_name) as fp:
        for line in fp:
            words = line.strip().split('\t')
            data_set.append([float(i) for i in words[:-1]])
            degrees.append(words[-1])
    degree_label = {degree: label for label, degree in enumerate(np.unique(degrees))}
    labels = [degree_label[i] for i in degrees]
    return np.array(data_set), np.array(labels).T
```

```python
def auto_morm(data_set):
    min_vals = np.min(data_set, axis=0)
    max_vals = np.max(data_set, axis=0)
    ranges = max_vals - min_vals
    m = np.shape(data_set)[0]
    norm_data_set = data_set - np.tile(min_vals, (m ,1))
    norm_data_set = norm_data_set / np.tile(ranges, (m, 1))
    return norm_data_set, ranges, min_vals
```

```python
norm_data_set, ranges, min_vals = auto_morm(dating_data_mat)
```

```python
def dating_class_test():
    ho_ratio = 0.10
    dating_data_max, dating_labels = file_to_matrix("datingTestSet.txt")
    norm_mat, ranges, min_vals = auto_morm(dating_data_mat)
    m = np.shape(norm_mat)[0]
    num_test_vecs = int(m*ho_ratio)
    error_count = 0.0
    for i in range(num_test_vecs):
        classifier_result = classify(norm_mat[i, :], norm_mat[num_test_vecs:m, :], dating_labels[num_test_vecs:m], 1)
        print("the classifier came back with: %s, the real answer is: %s" % (classifier_result, dating_labels[i]))
        if (classifier_result!=dating_labels[i]):
            error_count += 1.0
    print("the total error rate is: %f" % (error_count/float(num_test_vecs)))
```

```python
def img_to_vector(file_name):
    return_vect = []
    with open(file_name) as fp:
        for line in fp:
            return_vect.extend([int(c) for c in line.strip()])
    return np.array(return_vect, dtype=np.int64)
```

```python
import os
```

```python
def handwriting_class_test():
    data_set = []
    labels = []
    training_file_list = os.listdir("trainingDigits")
    for i in training_file_list:
        data_set.append(img_to_vector("trainingDigits/"+i))
        labels.append(int(i[0]))
    data_set = np.row_stack(data_set)
    labels = np.array(labels).T
    test_file_list = os.listdir("testDigits")
    error_count = 0.0
    m_test = len(test_file_list)
    for i in test_file_list:
        vector_under_test = img_to_vector("testDigits/"+i)
        classifier_result = classify(vector_under_test, data_set, labels, 3)
        print("the classifier came back with: %d, the real answer is: %d" % (classifier_result, int(i[0])))
        if classifier_result!=int(i[0]):
            error_count += 1.0
    print("The total error rate is: %f" % (error_count/m_test))    
```

```python
handwriting_class_test()
```
