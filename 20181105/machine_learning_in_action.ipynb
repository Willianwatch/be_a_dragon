```python
import numpy as np
```

```python
def load_data_set(file_name):
    return np.loadtxt(file_name, dtype=np.float64)
```

```python
def bin_split_data_set(data_set, feature, value):
    mat0 = data_set[data_set[:, feature]>value]
    mat1 = data_set[data_set[:, feature]<=value]
    return mat0, mat1
```

```python
def create_tree(data_set, leaf_type=reg_leaf, err_type=reg_err, ops=(1,4)):
    feat, val = choose_best_to_split(data_set, leaf_type, err_type, ops)
    if feat==None:
        return val
    ret_tree = {}
    ret_tree["sp_ind"] = feat
    ret_tree["sp_val"] = val
    l_set, r_set = bin_split_data_set(data_set, feat, val)
    ret_tree["left"] = create_tree(l_set, leaf_type, err_type, ops)
    ret_tree["right"] = create_tree(r_set, leaf_type, err_type, ops)
    return ret_tree
```

```python
def reg_leaf(data_set):
    return np.mean(data_set[:, -1])
```

```python
def reg_err(data_set):
    return np.var(data_set[:, -1])*np.shape(data_set)[0]
```

```python
def choose_best_to_split(data_set, leaf_type=reg_leaf, err_type=reg_err, ops=(1,4)):
    tol_s = ops[0]
    tol_n = ops[1]
    if len(set(data_set[:, -1]))==1:
        return None, leaf_type(data_set)
    m, n = np.shape(data_set)
    s = err_type(data_set)
    best_s = np.inf
    best_index = 0
    best_value = 0
    for feat_index in range(n-1):
        for split_val in set(data_set[:, feat_index]):
            mat0, mat1 = bin_split_data_set(data_set, feat_index, split_val)
            if (np.shape(mat0)[0]<tol_n) or (np.shape(mat1)[0]<tol_n):
                continue
            new_s = err_type(mat0) + err_type(mat1)
            if new_s<best_s:
                best_index = feat_index
                best_value = split_val
                best_s = new_s
    if (s-best_s)<tol_s:
        return None, leaf_type(data_set)
    mat0, mat1 = bin_split_data_set(data_set, best_index, best_value)
    if (np.shape(mat0)[0]<tol_n) or (np.shape(mat1)[0]<tol_n):
        return None, leaf_type
    return best_index, best_value
```

```python
def is_tree(obj):
    return type(obj).__name__=="dict"
```

```python
def get_mean(tree):
    if is_tree(tree["right"]):
        tree["right"] = get_mean(tree["right"])
    if is_tree(tree["left"]):
        tree["left"] = get_mean(tree["left"])
    return (tree["left"]+tree["right"])/2
```

```python
def prune(tree, test_data):
    if np.shape(test_data)[0]==0:
        return get_mean(tree)
    if is_tree(tree["right"]) or is_tree(tree["left"]):
        l_set, r_set = bin_split_data_set(test_data, tree["sp_ind"], tree["sp_val"])
    if is_tree(tree["left"]):
        tree["left"] = prune(tree["left"], l_set)
    if is_tree(tree["right"]):
        prune(tree["right"], r_set)
    if (not is_tree(tree["left"])) and (not is_tree(tree["right"])):
        l_set, r_set = bin_split_data_set(test_data, tree["sp_ind"], tree["sp_val"])
        error_no_merge = np.sum(np.power(l_set[:, -1]- tree["left"], 2)) + np.sum(np.power(r_set[:, -1] - tree["right"], 2))
        tree_mean = (tree["left"] + tree["right"]) / 2
        error_message = np.sum(np.power(test_data[:, -1] - tree_mean, 2))
        if error_message<error_message:
            print("merging")
            return tree_mean
        else:
            return tree
    else:
        return tree
```

```python
def linear_solve(data_set):
    m, n = np.shape(data_set)
    X = np.ones((m, n))
    Y = np.ones((m, 1))
    X[:, 1:n] = data_set[:, 0:n-1]
    Y = data_set[:, -1]
    xTx = np.dot(X.T, X)
    if np.linalg.det(xTx)==0.0:
        raise RuntimeError("This matrix is singular, cannot do inverse, try increasing the second value of ops")
    ws = np.dot(np.linalg.inv(xTx), np.dot(X.T, Y))
    return ws, X, Y
```

```python
def model_leaf(data_set):
    ws, X, Y = linear_solve(data_set)
    return ws
```

```python
def model_err(data_set):
    ws, X, Y = linear_solve(data_set)
    y_hat = np.dot(X, ws)
    return np.sum(np.power(Y-y_hat, 2))
```
